```python
def unir_pest(file_path):
    # Verificar las hojas presentes en el archivo
    xls = pd.ExcelFile(file_path)
    hojas_disponibles = xls.sheet_names
    
    # Inicializar una lista para almacenar los DataFrames de las hojas
    dataframes = []
    
    # Recorrer las hojas disponibles y leerlas
    for hoja in hojas_disponibles:
        # Leer el contenido de cada hoja
        df = pd.read_excel(file_path, sheet_name=hoja, skiprows=5)
        df['SHEET'] = hoja  # Añadir una columna con el nombre de la hoja
        dataframes.append(df)  # Agregar el DataFrame a la lista

    # Concatenar todos los DataFrames que se leyeron
    df_compras = pd.concat(dataframes, ignore_index=True)
    
    return df_compras
```
